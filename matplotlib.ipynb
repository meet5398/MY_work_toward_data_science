```python
from matplotlib import ticker#for making tick in(minor locator)
ax=plt.subplot()#declaring object
x=[12,23,45,56,56,56]
y=[34,45,54,76,87,89]
ax.plot(x,y,label='parul university')#giving label to the graph
ax.plot([23,434,434,23,43,3],label='india')
ax.set_title('my life my rules')#use to describe title
tick_spacing=50#definig space between sub tick
ax.yaxis.set_minor_locator(ticker.MultipleLocator(tick_spacing))
tick_spacing=5
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
#ax.yaxis.set_major_formatter('9')#it is used to formate major formatter
ax.set_xlabel('marks')
ax.set_ylabel('year')
plt.legend()#used to print labels
plt.show()
```
